```python
import numpy as np
import pandas as pd
from Querys.Query_geral import df_query_geral, din_tr, din_vcto_medio
from Querys.Analise_de_compra_por_estoque import df_analise_compra_por_estoque

# Removendo duplicatas
df_query_geral.drop_duplicates(subset='NOME', keep='first', inplace=True)

df_query_geral_reduzido = df_query_geral[['NOME', 'CODIGO', 'EMB_TRANSF']]

df_analise_reduzido = df_analise_compra_por_estoque['NOME']

# Realizando o merge dos DataFrames
df_completo = pd.merge(df_analise_reduzido, df_query_geral_reduzido, on='NOME', how='left')

df_completo
```

```python
# Reordenando as colunas
df_merged = pd.merge(df_completo, din_tr, on='CODIGO', how='left')

df_merged = pd.merge(df_merged, din_vcto_medio, on='CODIGO', how='left')

# Supondo que 'df' é o seu DataFrame e 'VCTO_MEDIO' é a coluna equivalente a G5
# Aplicando a operação e tratando erros (assumindo que os erros seriam divisões por zero ou dados não numéricos)
df_merged['VAL EM MESES'] = df_merged['VAL_MÉDIA'].apply(lambda x: np.floor(x / 30) * 0.5 if pd.notnull(x) else np.nan)

df_merged
```

```python
df_merged = pd.merge(df_merged, df_query_geral[['CODIGO', 'SETOR', 'GRUPO', 'SUBGRUPO']], on='CODIGO', how='left')

df_merged
```

```python
# Filtrando as colunas específicas de df_analise_compra_por_estoque
colunas_selecionadas = ['NOME', 'EST. ATUAL. SMJ', 'VDA MEDIA 30 D. SMJ', 'VDA 30D. SMJ', 'MAIOR VENDA SMJ', 
                        'est/vda * SMJ', 'PD. AB. SMJ', 'EST. ATUAL. STT', 'VDA MEDIA 30 D. STT', 'VDA 30D. STT', 
                        'MAIOR VENDA STT', 'est/vda * STT', 'PD. AB. STT']

df_analise_compra_por_estoque_filtrado = df_analise_compra_por_estoque[colunas_selecionadas]
df_analise_compra_por_estoque_filtrado = df_analise_compra_por_estoque_filtrado.drop_duplicates(subset='NOME', keep='first')

# Realizando o merge dos DataFrames com as colunas especificadas
df_merged = pd.merge(df_merged, df_analise_compra_por_estoque_filtrado, on='NOME', how='left')

df_merged
```

```python
# Usando o método apply com uma função lambda para aplicar a lógica condicional
df_merged['EST SMJ > X VDA'] = df_merged.apply(
    lambda row: "SIM" if not pd.isna(row['EST. ATUAL. SMJ']) and not pd.isna(row['VDA 30D. SMJ']) and row['EST. ATUAL. SMJ'] > (row['VDA 30D. SMJ'] * 1) else ("NÃO" if not pd.isna(row['EST. ATUAL. SMJ']) and not pd.isna(row['VDA 30D. SMJ']) else np.nan),
    axis=1
)

# Usando o método apply com uma função lambda para aplicar a lógica condicional
df_merged['EST SMJ > X EMB.'] = df_merged.apply(
    lambda row: "SIM" if not pd.isna(row['EST. ATUAL. SMJ']) and not pd.isna(row['EMB_TRANSF']) and row['EST. ATUAL. SMJ'] > (row['EMB_TRANSF'] * 1.5) else ("NÃO" if not pd.isna(row['EST. ATUAL. SMJ']) and not pd.isna(row['EMB_TRANSF']) else np.nan),
    axis=1
)

# Usando o método apply com uma função lambda para aplicar a lógica condicional
df_merged['EST STT > X VDA'] = df_merged.apply(
    lambda row: "SIM" if not pd.isna(row['EST. ATUAL. STT']) and not pd.isna(row['MAIOR VENDA STT']) and row['EST. ATUAL. STT'] > (row['MAIOR VENDA STT'] * 2) else ("NÃO" if not pd.isna(row['EST. ATUAL. STT']) and not pd.isna(row['MAIOR VENDA STT']) else np.nan),
    axis=1
)

df_merged
```

```python
def custom_formula(row):
    try:
        if row['TR VIX'] == 1:
            if row['MAIOR VENDA SMJ'] == 0:
                result = np.floor(0.20 * row['EST. ATUAL. SMJ'] / row['EMB_TRANSF']) * row['EMB_TRANSF']
            else:
                temp_result = np.floor((row['EST. ATUAL. SMJ'] - (1.2 * row['MAIOR VENDA SMJ'])) / row['EMB_TRANSF']) * row['EMB_TRANSF']
                result = max(0, temp_result)
            return result * 0.50  # Multiplica por 0.5 se 'TR VIX' é 1
        else:
            if row['MAIOR VENDA SMJ'] == 0:
                result = np.floor(0.20 * row['EST. ATUAL. SMJ'] / row['EMB_TRANSF']) * row['EMB_TRANSF']
            else:
                temp_result = np.floor((row['EST. ATUAL. SMJ'] - (1.2 * row['MAIOR VENDA SMJ'])) / row['EMB_TRANSF']) * row['EMB_TRANSF']
                result = max(0, temp_result)
            return result  # Retorna o resultado sem multiplicar por 0.5
    except Exception as e:
        return np.nan  # Retorna NaN em caso de erro

df_merged['EXCEDENTE SMJ'] = df_merged.apply(custom_formula, axis=1)

df_merged
```

```python
def custom_formula(row):
    try:
        # Verifica se EMB_TRANSF é 0 e retorna 'EXCEDENTE SMJ' se for o caso
        if row['MAIOR VENDA STT'] == 0:
            return row['EXCEDENTE SMJ']  # Corresponde ao Z1170 no Excel

        # Calcula o valor e garante que seja no mínimo 0
        valor_calculado = np.ceil(((row['MAIOR VENDA STT'] * min(row['VAL EM MESES'], 2)) - row['EST. ATUAL. STT']) / row['EMB_TRANSF']) * row['EMB_TRANSF']
        return max(valor_calculado, 0)
    except:
        return np.nan  # Retorna "" em caso de qualquer outro erro

# Aplicando a função ao DataFrame
df_merged.loc[:, 'NECESSIDADE STT'] = df_merged.apply(custom_formula, axis=1)
df_merged.loc[:, 'NECESSIDADE STT'] = df_merged['NECESSIDADE STT'].astype('float64')

df_merged
```

```python
# Realizando as conversões
df_merged.loc[:, 'NECESSIDADE STT'] = pd.to_numeric(df_merged['NECESSIDADE STT'], errors='coerce')
df_merged.loc[:, 'EXCEDENTE SMJ'] = pd.to_numeric(df_merged['EXCEDENTE SMJ'], errors='coerce')

# Aplicando a função
df_merged.loc[:, 'PEDIDO'] = df_merged.apply(lambda row: row['EXCEDENTE SMJ'] if row['NECESSIDADE STT'] > row['EXCEDENTE SMJ'] else row['NECESSIDADE STT'], axis=1)

df_merged
```

```python
# Aplicando os filtros
df_filtrado = df_merged[
    (df_merged["TR SMJ"] == 0) &
    (df_merged["TR STT"] == 1) &
    (df_merged['EST SMJ > X VDA'] == 'SIM') &
    (df_merged['EST SMJ > X EMB.'] == 'SIM') &
    (df_merged['EST STT > X VDA'] == 'NÃO') &
    (df_merged['EXCEDENTE SMJ'] > 0) &
    (df_merged['NECESSIDADE STT'] > 0) &
    (df_merged['PEDIDO'] > 0)
]

df_final = df_filtrado.copy()
df_final['PEDIDO'] = df_final['PEDIDO'].fillna(0).astype('int64')

df_final
```

```python
df_csv = df_final[['CODIGO', 'PEDIDO']]
df_csv
```
